```python
from PIL import Image

targetX=100
targetY=150
print('\nDownscaling images for traning...')

for i in tqdm(dataTraining.index):
    tempImage = Image.open(os.path.join('images/', str(i) + '.jpeg'))
    tempImage = tempImage.resize((targetX,targetY),Image.ANTIALIAS)
    tempImage.save(os.path.join('images_resize_color/', str(i) + '.jpeg'),optimize=True,quality=95)
    

print('\nDownscaling images for testing...')

for i in tqdm(dataTesting.index):
    tempImage = Image.open(os.path.join('images/', str(i) + '.jpeg'))
    tempImage = tempImage.resize((targetX,targetY),Image.ANTIALIAS)
    tempImage.save(os.path.join('images_resize_color/', str(i) + '.jpeg'),optimize=True,quality=95)
```
